## Load 1-Day NO2 DataFrames as GeoDataFrames

We convert the Pandas NO2 DataFrames created by `nc_to_df` into GeoPandas GeoDataFrames using the `latitude` and `longitude` columns.

### Swath 1

```python
gdf1 = gpd.GeoDataFrame(df1, geometry=gpd.points_from_xy(df1.longitude, df1.latitude),crs=crs)
```

```python
gdf1_sjoin = gpd.sjoin(gdf1, africa_simple_gdf, how='inner', op='intersects')
```

```python
fig, ax = plt.subplots(figsize=(15,10))
africa_simple_gdf.plot(cmap=plt.get_cmap('gray'),ax=ax, alpha=0.1)
gdf1_sjoin.plot(cmap=plt.get_cmap('Reds'), ax=ax, column='qa_value',legend=True)
```
